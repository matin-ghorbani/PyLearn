## Import Dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from lls import LLS
from scipy.stats import linregress
```

## Generate continuous random dataset for the problem of X = study hours and Y = grade of students (if you study more, you will get a higher grade).

```python
X = np.random.uniform(0, 10, 300)
Y = (2. * X) + (np.random.normal(0, 1, size=X.shape))

X, Y = X.reshape(-1, 1), Y.reshape(-1, 1)
```

## Show Random dates

```python
plt.scatter(X, Y)
plt.title('Random Data Of Hours And Grades')
plt.xlabel('hours')
plt.ylabel('grade')
plt.show()
```

## Split your dataset to train and test with train_test_split function.

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=.2)
```

## Draw some wrong red lines with a random slope.

```python
plt.scatter(X, Y)
plt.plot(np.linspace(0, 10, 20), np.linspace(0, 10, 20), color="red")
plt.title('Random Data Of Hours And Grades With A Random Slope')
plt.xlabel('hours')
plt.ylabel('grade')
plt.show()
```

## Draw the correct blue line using your Object Oriented Linear Least Squares (LLS) method.

```python
lls = LLS()
lls.fit(np.array(x_train), np.array(y_train))
w = lls.w

plt.scatter(X, Y)
plt.plot(X, X * w, color='blue')
plt.show()
```

## Draw a green line using linregress from scipy library. Compare your result with scipy's result.

```python
X, Y = x_train, y_train
X, Y = X.reshape(-1), Y.reshape(-1)


slope, intercept, r, p, std_err = linregress(X, Y)

mymodel = list(map(lambda x: slope * x + intercept, x_train))

plt.scatter(x_train, y_train)
plt.plot(x_train, mymodel, color='green')
plt.show()
```
