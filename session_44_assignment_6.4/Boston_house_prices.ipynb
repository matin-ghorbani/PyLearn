## Import Dependencies

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from lls import LLS
```

## Load Boston dataset

```python
data = pd.read_csv('Boston.csv')
data = data.drop('Unnamed: 0', axis=1)
data.rename(columns={'medv': 'price'}, inplace=True)

data.head()
```

## Find out which 2 xes are better

```python
data.corr()
```

### rm has the better with .6953 result and then zn with .3604

```python
X = np.array([data['rm'], data['zn']])
X = X.T

Y = np.array(data['price'])

X
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=.2)
```

## Solve the regression problem on 2 dimensions (2 features) using our Object Oriented LLS.

```python
lls = LLS()
lls.fit(x_train, y_train)

w = lls.w
w
```

## Show results on a 3D plot and draw the y_pred as a plane on it.

```python
X1, X2 = x_train[:,0], x_train[:,1]
x1, x2 = np.meshgrid(X1, X2)
y_pred1, y_pred2 = w[0] * x1, w[1] * x2

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X1 , X2 , y_train,  color='aqua',alpha=0.3)
ax.plot_surface(x1, x2,  y_pred1 + y_pred2)
ax.set_xlabel('rm')
ax.set_ylabel('zn')
ax.set_zlabel('price')
plt.title('Results Plot')
plt.show()
```
