## Import Dependencies

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from perceptron import Perceptron
```

## Make a linear dataset using scikit-learn library

```python
x, y, coef = make_regression(n_samples=100,
                             n_features=1,
                             n_informative=1,
                             noise=10,
                             coef=True,
                             random_state=0)

# Scale feature x (years of experience) to range 0..20
x = np.interp(x, (x.min(), x.max()), (0, 20))

# Scale target y (salary) to range 20000..150000 
y = np.interp(y, (y.min(), y.max()), (20000, 150000))

plt.ion()
plt.plot(x, y, '.', label='training data')
plt.xlabel('Years of experience')
plt.ylabel('Salary $')
plt.title('Experience And Salary')
```

## Split data to train and test datasets

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2)
```

## Implement the perceptron algorithm as a class and fit our model on the employee's salary dataset

```python
model = Perceptron(lr_w=.0001, lr_b=.1, epochs=3)
model.fit(x_train, y_train)
y_pred = model.predict(x_train)
```

## Plot data graph and loss graph as 2 subplots in 1 window

```python
(ax1, ax2) = plt.subplots(1, 2)[1]
ax1.scatter(x_train, y_train, color='blue')
ax1.plot(x_train, y_pred, color='red')
ax1.set_xlabel('Years')
ax1.set_ylabel('Salary')
ax1.set_title('Experience And Salary')

ax2.plot(model.losses, color='brown')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Loss')
ax2.set_title('Loss And Epochs With 3 Epochs')
plt.show()
```

## Change hyperparameters for get best result

```python
model = Perceptron(lr_w=.0001, lr_b=.1, epochs=20)
model.fit(x_train, y_train)
y_pred = model.predict(x_train)
```
